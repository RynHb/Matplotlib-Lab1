```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
lst = ['Ahmed','Khalid','Ali','Rayan','Ahmed']
lst2 = [12000,8000,25000,10000,6000]
df_data = {'Sport_name':['Water','Football','Basketball'],
            'Team_name':['Ittihad','Hilal','Nasr'],
            'City':['Jeddah','Riyadh','Riyadh'],
            'Profit':[150000,300000,500000]}
df = pd.DataFrame(df_data)
```

```python
plt.bar(lst,lst2)
plt.title('Employees Salaries', fontsize = 20)
plt.xlabel('Names')
plt.ylabel('Salary')
plt.show()
```

```python
plt.hist(lst)
plt.title('Names Frequncey', fontsize = 20)
plt.xlabel('Names')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(df.Team_name,df.City)
plt.title('Teams City', fontsize=15)
plt.xlabel('Teams')
plt.ylabel('City')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(df.Profit, labels=df.Team_name, autopct='%1.2f%%', explode=[0.1,0.1,0.1])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df.Team_name, df.Profit, marker = 'o', alpha=0.4, s=50)
plt.title('Scatter of Teams Profit')
plt.xlabel('Teams')
plt.ylabel('Profit')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df.Team_name, df.Profit, marker = 'o', alpha=0.4, s=50)
plt.title('Scatter of Teams Profit', loc='left', color='r')
plt.xlabel('Teams', fontsize=16)
plt.ylabel('Profit', fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot([1,7,9,20,55,70], 'r')
plt.plot([7.5,15.4,9.5,10.7], 'g')
plt.plot([44,50,33,12,7,10], 'b')
plt.legend(['Teams Numbers','Ratio Numbers','City Numbers'], shadow=True, loc=0)
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=[20,10])
plt.suptitle('Sport Teams Charts', fontsize=25)

plt.subplot(3,1,1)
plt.plot([1,7,9,20,55,70], 'r')
plt.plot([7.5,15.4,9.5,10.7], 'g')
plt.plot([44,50,33,12,7,10], 'b')
plt.legend(['Teams Numbers','Ratio Numbers','City Numbers'], shadow=True, loc=0);

plt.subplot(3,1,2)
plt.scatter(df.Team_name, df.Profit, marker = 'o', alpha=0.4, s=50)
plt.title('Scatter of Teams Profit', loc='left', color='r')
plt.xlabel('Teams', fontsize=16)
plt.ylabel('Profit', fontsize=16);

plt.subplot(3,1,3)
plt.pie(df.Profit, labels=df.Team_name, autopct='%1.2f%%', explode=[0.1,0.1,0.1]);
```
